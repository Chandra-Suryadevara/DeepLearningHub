***Classification Metrics***

Different Classification Metrics
1. Accuracy
2. Recall
3. Precision
4. F1-Score

***Accuracy*** in classification problems is the number of correct predictions made by the model divided by the total number of predictions

Example:
Model predicted 80 correct images out of 100 then 80/100 == 0.8 or 80% accuracy

Useful Situations:
1. Very good at balanced class means out of 200 images 100 are dogs and 100 are cats which is balanced


***Recall*** Ability of a model to find all the relevant cases within dataset 

No.of true positive // No.of positive + No.of Negative




***True Positives (TP)***: These are instances where the model correctly predicts the positive class. For example, if the task is to identify emails as spam, a true positive is an email that is indeed spam and is correctly identified as such by the model.

***False Negatives (FN)***: These are instances where the model fails to predict the positive class. In the email example, a false negative would be an email that is spam but is incorrectly identified as not spam by the model.

***Precision*** Ability of the classification model to identify only the relevant data points. 


* Precision is defined as the number of true positives divided by the number of true positives plus the number of false positives

To understand this definition, let's break down the components:

***True Positives (TP)***: These are instances where the model correctly predicts the positive class. For example, if the task is to identify emails as spam, a true positive is an email that is indeed spam and is correctly identified as such by the model.

***False Positives (FP)***: These are instances where the model incorrectly predicts the positive class. Continuing with the email example, a false positive would be an email that is not spam (it's a legitimate email) but is incorrectly identified as spam by the model.

***F1 Score***: Optimal blend of precision and recall

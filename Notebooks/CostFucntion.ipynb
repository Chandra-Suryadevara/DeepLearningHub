### Cost (Loss) Functions in Neural Networks




#### 1. Mean Squared Error (MSE)
- **Formula:**
  \[
  \text{MSE} = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2
  \]
  where \(y_i\) is the true value and \(\hat{y}_i\) is the predicted value.
- **Range:** [0, ∞)
- **Properties:**
  - **Used for Regression:** Suitable for continuous output predictions.
  - **Sensitive to Outliers:** Large errors are penalized more heavily.

#### 2. Mean Absolute Error (MAE)
- **Formula:**
  \[
  \text{MAE} = \frac{1}{n} \sum_{i=1}^n |y_i - \hat{y}_i|
  \]
- **Range:** [0, ∞)
- **Properties:**
  - **Used for Regression:** Suitable for continuous output predictions.
  - **Less Sensitive to Outliers:** Errors are weighted linearly.

#### 3. Cross-Entropy Loss (Log Loss)
- **Binary Classification Formula:**
  \[
  \text{CE} = -\frac{1}{n} \sum_{i=1}^n [y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i)]
  \]
- **Multi-class Classification Formula:**
  \[
  \text{CE} = -\frac{1}{n} \sum_{i=1}^n \sum_{j=1}^c y_{ij} \log(\hat{y}_{ij})
  \]
  where \(c\) is the number of classes.
- **Range:** [0, ∞)
- **Properties:**
  - **Used for Classification:** Suitable for both binary and multi-class problems.
  - **Probabilistic Interpretation:** Measures the difference between two probability distributions.
